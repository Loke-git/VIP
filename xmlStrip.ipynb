# Testing

```python
xmlpath = "src_xml\plant_catalog.xml" # Definer PATH til testxml
with open(xmlpath, "r") as file:
    # Read each line in the file, readlines() returns a list of lines
    content = file.readlines()
    # Combine the lines in the list into a string
    content = "".join(content)
    bs_content = BeautifulSoup(content, "lxml")
```

```python
#bs_content #Vis XML
result = bs_content.find_all("zone") #Finn streng i xml-koding (ikke i innhold)
#Finner kun 1:1 verbatim - mangler wildcard eller?
result # Vis result
```

```python
# Finn ALLE tags i dokumentet via etree
import xml.etree.ElementTree as ET
# load and parse the file
xmlTree = ET.parse(xmlpath)

elemList = []

for elem in xmlTree.iter():
    elemList.append(elem.tag)

# Fjern duplikater ved å konvertere til sett og tilbake til liste.
# Vi kunne ha sjekket unikhet tidligere - men dette er én prosess fremfor flere tusen. Sparer kanskje et par millisekunder.
elemList = list(set(elemList))

# Just printing out the result
print(elemList)
```

## Total taggfjerning via RegEx

Dette steget utrydder alle tagger uten å berøre innhold. Dette bør være det **siste** steget vi tar for å fjerne tagger, altså ETTER spesielle tagger hvis innhold må fjernes er borte.

*Huskeliste: del var xmlpath i PATH og filnavn*

```python
# Ordinær RegEx-vask
import re
# For each...
text = re.sub('<[^<]+>', "", open(xmlpath).read())
name = xmlpath+"_SANERT" # Med xmlpath blir denne stappet i original path. Oi da
with open(name+".txt", "w") as f:
    f.write(text)
```
